```python
import pylab as plt

def show(x, y, xleg, yleg):
    plt.plot(x, y)
    plt.yscale('log')
    plt.xlabel(xleg, fontsize=17)
    plt.ylabel(yleg, fontsize=17)
    
    
```

```python
plt.yscale('log')
x = [i for i in range(4, 65, 4)]

f = lambda x: x*x*2**x
y = [f(i) for i in x]
plt.plot(x, y, 'r')

f = lambda x: x*2**(x-1)
y = [f(i) for i in x]
plt.plot(x, y, 'b')

f = lambda x: 2**(x-3)
y = [f(i) for i in x]
plt.plot(x, y, 'g')

f = lambda x: x**2 * 100000
y = [f(i) for i in x]
plt.plot(x, y, 'y')

plt.legend(["O(NN2^N)", "O(N2^(N-1))", "O(2^(N-3))", "O(NNM), M=100000"], loc=0)
plt.xlabel('N', fontsize=17)
plt.ylabel('Оценка Сложности', fontsize=17)
plt.show()
```
