```python
import math
import numpy as np

def plus(a, b):
    return [a[0] + b[0], a[1] + b[1]]

def minus(a, b):
    return [a[0] - b[1], a[1] - b[0]]

def mult(a, b):
    min1 = min(a[0] * b[0], a[0] * b[1], a[1] * b[0], a[1] * b[1])
    min2 = max(a[0] * b[0], a[0] * b[1], a[1] * b[0], a[1] * b[1])
    return [min1, min2]

def div(a, b):
    min1 = min(a[0] / b[0], a[0] / b[1], a[1] / b[0], a[1] / b[1])
    min2 = max(a[0] / b[0], a[0] / b[1], a[1] / b[0], a[1] / b[1])
    return [min1, min2]

a = [[[1.2,1.2] ,[0, 1]] , [[-1.2, -1.2] , [1,0]]]
b = [[2,2], [2, 2]]
#print('2x2 matrix is:\n', a)
#print('1x2 matrix is:\n', b)

for i in range(len(a[0])):
    for j in range(len(a[0])):
        a[i][j] = div(a[i][j], b[i])
        
ans = [a[0][0], a[1][0]] 

print(ans)

temp_ans = ans

while(1):
    for i in range(len(a[0])):
        temp_ans[i] = minus(a[i][0], mult(a[i][1] , ans[i]))
        if((temp_ans[0][0] - ans[0][0])**2 < 0.1 and (temp_ans[1][1] - ans[1][1])**2 < 0.1):
            ans = temp_ans
        ans = temp_ans
        print(ans)

print(ans)
```
