```python

import matplotlib.pyplot as plt
import numpy as np

def feval(funcName, *args):
    return eval(funcName)(*args)

def RK4(func, yinit, x_range, h):
    m = len(yinit)
    n = int((x_range[-1] - x_range[0])/h)

    x = x_range[0]
    y = yinit

    # Containers for solutions
    xsol = np.empty(0)
    xsol = np.append(xsol, x)

    ysol = np.empty(0)
    ysol = np.append(ysol, y)

    for i in range(n):
        k1 = feval(func, x, y)
        yp2 = y + k1*(h/2)
        k2 = feval(func, x+h/2, yp2)
        yp3 = y + k2*(h/2)
        k3 = feval(func, x+h/2, yp3)
        yp4 = y + k3*h
        k4 = feval(func, x+h, yp4)

        for j in range(m):
            y[j] = y[j] + (h/6)*(k1[j] + 2*k2[j] + 2*k3[j] + k4[j])

        x += h
        xsol = np.append(xsol, x)

        for r in range(len(y)):
            ysol = np.append(ysol, y[r])

    return [xsol, ysol]

def ODES(x, y):
    al1 = 1.5
    al2 = 1
    g1 = 0.8
    g2 = 0.4
    dy = np.zeros((len(y)))
    dy[0] = -y[0] * al1 + g1 * y[0] * y[1]
    dy[1] = y[1] * al2 - g2 * y[0] * y[1]
    return dy

h = 0.1
x = np.array([0.0, 40.0])
yinit = np.array([10.0, 2.0])

[ts, ys] = RK4('ODES', yinit, x, h)

node = len(yinit)
ys1 = ys[0::node]
ys2 = ys[1::node]

plt.plot(ts, ys1, 'r')
plt.plot(ts, ys2, 'b')
plt.xlim(x[0], x[1])
plt.legend(["u", "v"], loc=2)
plt.xlabel('t', fontsize=17)
plt.ylabel('u/v', fontsize=17)
plt.tight_layout()
plt.show()

plt.plot(ys2, ys1, 'g')
plt.xlabel('v', fontsize=17)
plt.ylabel('u', fontsize=17)
plt.tight_layout()
plt.show()
```

```python
def trapezoidal(f, a, b, n):
    h = (b - a)/n
    result = 0.5*(f[0] + f[n - 1])
    for i in range(1, n):
        result += f[i]
    result *= h
    return result
```

```python
def solverAdv(al1, al2, g1 , g2 , u_0, v_0, u_exp, v_exp, t_exp):
    #print(al1, al2)
    #h = 0.1
    #x = np.array([0.0, 40.0])
    #yinit = np.array([u_0, v_0])

    #[ts, ys] = RK4('ODES', yinit, x, h)
    #node = len(yinit)
    #ys1 = ys[0::node]
    #ys2 = ys[1::node]
    
    t_calc = 2 * math.pi / math.sqrt(al1 * al2)
    #u_calc = 1 / t_calc * trapezoidal(ys1, 0, t_calc, ys1.size)
    #v_calc = 1 / t_calc * trapezoidal(ys2, 0, t_calc, ys2.size)
    
    u_calc = al2 / g2
    v_calc = al1 / g1
    
    #print("U")
    #print(u_calc,u_exp)
    #print("V")
    #print(v_calc,v_exp)
    #print("T")
    #print(t_calc,t_exp)
    #print(((u_calc - u_exp) / u_exp) ** 2 + ((v_calc - v_exp) / v_exp) ** 2 + ((t_calc - t_exp) / t_exp) ** 2)
    
    #return ((u_exp - u_calc) / u_calc) ** 2 + ((v_exp - v_calc) / v_calc) ** 2 + ((t_exp - t_calc) / t_calc) ** 2
    return ((u_calc - u_exp) / u_exp) ** 2 + ((v_calc - v_exp) / v_exp) ** 2 + ((t_calc - t_exp) / t_exp) ** 2
```

```python
def disp(p1, p2, p3, pp1, pp2, pp3):
    calc1 = ((p1[0] + p2[0] + p3[0]) / 3,(p1[1] + p2[1] + p3[1]) / 3)
    calc2 = ((pp1[0] + pp2[0] + pp3[0]) / 3,(pp1[1] + pp2[1] + pp3[1]) / 3)
    if((calc1[0] - calc2[0])**2 < 0.001 and (calc1[1] - calc2[1])**2 < 0.001):
        return 0
    else:
        return 1
```

```python
def disp2(p1, p2, p3):
    f1 = solverF(p1[0], p1[1], g1, g2, u_avg, v_avg ,t_avg)
    f2 = solverF(p2[0], p2[1], g1, g2, u_avg, v_avg ,t_avg)
    f3 = solverF(p3[0], p3[1], g1, g2, u_avg, v_avg ,t_avg)
    if(((f1 - f2) / 2)**2 + ((f1 - f3) / 2)**2 + ((f2 - f3) / 2)**2 < 0.001):
        return 0
    else:
        return 1
```

```python
def disp3(p1, p2, p3):
    f1 = solverF(p1[0], p1[1], g1, g2, u_avg, v_avg ,t_avg)
    f2 = solverF(p2[0], p2[1], g1, g2, u_avg, v_avg ,t_avg)
    f3 = solverF(p3[0], p3[1], g1, g2, u_avg, v_avg ,t_avg)
    f_avg = (f1 + f2 + f3) / 3
    calc = ((f1 - f_avg)**2) / 3 + ((f2 - f_avg)**2) / 3 + ((f3 - f_avg)**2) / 3
    #print(math.sqrt(calc))
    if(math.sqrt(calc) < 0.000001):
        return 0
    else:
        return 1
```

```python
def disp4(p1, p2, p3):
    f1 = solverAdv(p1[0], p1[1], g1, g2, u_0, v_0, u_avg, v_avg ,t_avg)
    f2 = solverAdv(p2[0], p2[1], g1, g2, u_0, v_0, u_avg, v_avg ,t_avg)
    f3 = solverAdv(p3[0], p3[1], g1, g2, u_0, v_0, u_avg, v_avg ,t_avg)
    if(((f1 - f2) / 2)**2 + ((f1 - f3) / 2)**2 + ((f2 - f3) / 2)**2 < 0.001):
        return 0
    else:
        return 1
```

```python
def disp5(p1, p2, p3):
    f1 = solverAdv(p1[0], p1[1], g1, g2, u_0, v_0, u_avg, v_avg ,t_avg)
    f2 = solverAdv(p2[0], p2[1], g1, g2, u_0, v_0, u_avg, v_avg ,t_avg)
    f3 = solverAdv(p3[0], p3[1], g1, g2, u_0, v_0, u_avg, v_avg ,t_avg)
    f_avg = (f1 + f2 + f3) / 3
    calc = ((f1 - f_avg)**2) / 3 + ((f2 - f_avg)**2) / 3 + ((f3 - f_avg)**2) / 3
    #print(math.sqrt(calc))
    if(math.sqrt(calc) < 0.001):
        return 0
    else:
        return 1
```

```python
def disp6(p1, p2, p3):
    #calc = (p1[0] + p2[0] + p3[0])**2 - (p1[1] + p2[1] + p3[1])**2
    calc1 = (p1[0] - p2[0]) + (p1[0] - p3[0]) + (p2[0] - p3[0])
    calc2 = (p1[1] - p2[1]) + (p1[1] - p3[1]) + (p2[1] - p3[1])
    calc = (calc1 + calc2)**2
    #print("Disp")
    #print(math.sqrt(calc))
    if(math.sqrt(calc) < 0.000001):
        return 0
    else:
        return 1
```

```python
def BlackBox (p1, p2, p3):
    #print(sortedF)
    sortedF = sorted([solverF(*p, g1, g2, u_avg, v_avg ,t_avg),p] for p in (p1, p2, p3)) # wild Python magic
    
    #print(sortedF)
    
    centroid = ((sortedF[0][1][0] + sortedF[1][1][0]) / 2,(sortedF[0][1][1] + sortedF[1][1][1]) / 2)
    x_r = (2 * centroid[0] - sortedF[2][1][0],2 * centroid[1] - sortedF[2][1][1]) # alpha is 1.0 
    fr = solverF(x_r[0], x_r[1], g1, g2, u_avg, v_avg ,t_avg)
    
    if(fr < sortedF[0][0]):
        x_e = (-1.0 * centroid[0] + 2.0 * x_r[0],-1.0 * centroid[1] + 2.0 * x_r[1]) # gamma is 2
        fe = solverF(x_e[0], x_e[1], g1, g2, u_avg, v_avg ,t_avg)
        if(fe < sortedF[0][0]):
            sortedF[2][1] = x_e
        else:
            sortedF[2][1] = x_r
    elif(fr > sortedF[0][0] and fr < sortedF[1][0]):
        sortedF[2][1] = x_r
    elif(fr > sortedF[1][0] and fr < sortedF[2][0]):
        sortedF[2][0] = fr
        sortedF[2][1] = x_r
        
        x_s = (0.5 * sortedF[2][1][0] + 0.5 * centroid[0], 0.5 * sortedF[2][1][1] + 0.5 * centroid[1])# beta is 0.5
        fs = solverF(x_s[0], x_s[1], g1, g2, u_avg, v_avg ,t_avg)
        
        if(fs < sortedF[2][0]):
            sortedF[2][1] = x_s
        else:
            #sortedF[1][1] = (sortedF[0][1][0] + (sortedF[1][1][0] - sortedF[0][1][0]) / 2, sortedF[0][1][1] + (sortedF[1][1][1] - sortedF[0][1][1]) / 2)
            #sortedF[2][1] = (sortedF[0][1][0] + (sortedF[2][1][0] - sortedF[0][1][0]) / 2, sortedF[0][1][1] + (sortedF[2][1][1] - sortedF[0][1][1]) / 2)
            sortedF[0][1] = ((sortedF[0][1][0] + sortedF[0][1][0]) / 2, (sortedF[0][1][1] + sortedF[0][1][1]) / 2)
            sortedF[1][1] = ((sortedF[1][1][0] + sortedF[0][1][0]) / 2, (sortedF[1][1][1] + sortedF[0][1][1]) / 2)
            sortedF[2][1] = ((sortedF[2][1][0] + sortedF[0][1][0]) / 2, (sortedF[2][1][1] + sortedF[0][1][1]) / 2)
    elif(fr > sortedF[2][0]):  
        x_s = (0.5 * sortedF[2][1][0] + 0.5 * centroid[0], 0.5 * sortedF[2][1][1] + 0.5 * centroid[1])# beta is 0.5
        fs = solverF(x_s[0], x_s[1], g1, g2, u_avg, v_avg ,t_avg)
        
        if(fs < sortedF[2][0]):
            sortedF[2][1] = x_s
        else:
            #sortedF[1][1] = (sortedF[0][1][0] + (sortedF[1][1][0] - sortedF[0][1][0]) / 2, sortedF[0][1][1] + (sortedF[1][1][1] - sortedF[0][1][1]) / 2)
            #sortedF[2][1] = (sortedF[0][1][0] + (sortedF[2][1][0] - sortedF[0][1][0]) / 2, sortedF[0][1][1] + (sortedF[2][1][1] - sortedF[0][1][1]) / 2)
            sortedF[0][1] = ((sortedF[0][1][0] + sortedF[0][1][0]) / 2, (sortedF[0][1][1] + sortedF[0][1][1]) / 2)
            sortedF[1][1] = ((sortedF[1][1][0] + sortedF[0][1][0]) / 2, (sortedF[1][1][1] + sortedF[0][1][1]) / 2)
            sortedF[2][1] = ((sortedF[2][1][0] + sortedF[0][1][0]) / 2, (sortedF[2][1][1] + sortedF[0][1][1]) / 2)
    points = [sortedF[0][1], sortedF[1][1], sortedF[2][1]]
    return points
```

```python
def WhiteBox (p1, p2, p3):
    sortedF = sorted([solverAdv(*p, g1, g2, u_0, v_0, u_avg, v_avg ,t_avg),p] for p in (p1, p2, p3)) # wild Python magic  
    
    #print(sortedF)
    
    centroid = ((sortedF[0][1][0] + sortedF[1][1][0]) / 2,(sortedF[0][1][1] + sortedF[1][1][1]) / 2)
    x_r = [2 * centroid[0] - sortedF[2][1][0],2 * centroid[1] - sortedF[2][1][1]] # alpha is 1.0 
    if(x_r[0] <= 0):
        x_r[0] = 0.001
    if(x_r[1] <= 0):
        x_r[1] = 0.001
    fr = solverAdv(x_r[0], x_r[1], g1, g2, u_0, v_0, u_avg, v_avg ,t_avg)
    
    if(fr < sortedF[0][0]):
        x_e = [-1.0 * centroid[0] + 2.0 * x_r[0],-1.0 * centroid[1] + 2.0 * x_r[1]] # gamma is 2
        if(x_e[0] <= 0):
            x_e[0] = 0.001
        if(x_e[1] <= 0):
            x_e[1] = 0.001
        fe = solverAdv(x_e[0], x_e[1], g1, g2, u_0, v_0, u_avg, v_avg ,t_avg)
        if(fe < sortedF[0][0]):
            sortedF[2][1] = x_e
        else:
            sortedF[2][1] = x_r
    elif(fr > sortedF[0][0] and fr < sortedF[1][0]):
        sortedF[2][1] = x_r
    elif(fr > sortedF[1][0] and fr < sortedF[2][0]):
        sortedF[2][0] = fr
        sortedF[2][1] = x_r
        
        x_s = (0.5 * sortedF[2][1][0] + 0.5 * centroid[0], 0.5 * sortedF[2][1][1] + 0.5 * centroid[1])# beta is 0.5
        if(x_s[0] <= 0):
            x_s[0] = 0.001
        if(x_s[1] <= 0):
            x_s[1] = 0.001
        fs = solverAdv(x_s[0], x_s[1], g1, g2, u_0, v_0, u_avg, v_avg ,t_avg)
        
        if(fs < sortedF[2][0]):
            sortedF[2][1] = x_s
        else:
            #sortedF[1][1] = (sortedF[0][1][0] + (sortedF[1][1][0] - sortedF[0][1][0]) / 2, sortedF[0][1][1] + (sortedF[1][1][1] - sortedF[0][1][1]) / 2)
            #sortedF[2][1] = (sortedF[0][1][0] + (sortedF[2][1][0] - sortedF[0][1][0]) / 2, sortedF[0][1][1] + (sortedF[2][1][1] - sortedF[0][1][1]) / 2)
            sortedF[0][1] = ((sortedF[0][1][0] + sortedF[0][1][0]) / 2, (sortedF[0][1][1] + sortedF[0][1][1]) / 2)
            sortedF[1][1] = ((sortedF[1][1][0] + sortedF[0][1][0]) / 2, (sortedF[1][1][1] + sortedF[0][1][1]) / 2)
            sortedF[2][1] = ((sortedF[2][1][0] + sortedF[0][1][0]) / 2, (sortedF[2][1][1] + sortedF[0][1][1]) / 2)
    elif(fr > sortedF[2][0]):  
        x_s = (0.5 * sortedF[2][1][0] + 0.5 * centroid[0], 0.5 * sortedF[2][1][1] + 0.5 * centroid[1])# beta is 0.5
        if(x_s[0] <= 0):
            x_s[0] = 0.001
        if(x_s[1] <= 0):
            x_s[1] = 0.001
        fs = solverAdv(x_s[0], x_s[1], g1, g2, u_0, v_0, u_avg, v_avg ,t_avg)
        
        if(fs < sortedF[2][0]):
            sortedF[2][1] = x_s
        else:
            #sortedF[1][1] = (sortedF[0][1][0] + (sortedF[1][1][0] - sortedF[0][1][0]) / 2, sortedF[0][1][1] + (sortedF[1][1][1] - sortedF[0][1][1]) / 2)
            #sortedF[2][1] = (sortedF[0][1][0] + (sortedF[2][1][0] - sortedF[0][1][0]) / 2, sortedF[0][1][1] + (sortedF[2][1][1] - sortedF[0][1][1]) / 2)
            sortedF[0][1] = ((sortedF[0][1][0] + sortedF[0][1][0]) / 2, (sortedF[0][1][1] + sortedF[0][1][1]) / 2)
            sortedF[1][1] = ((sortedF[1][1][0] + sortedF[0][1][0]) / 2, (sortedF[1][1][1] + sortedF[0][1][1]) / 2)
            sortedF[2][1] = ((sortedF[2][1][0] + sortedF[0][1][0]) / 2, (sortedF[2][1][1] + sortedF[0][1][1]) / 2)
    points = [sortedF[0][1], sortedF[1][1], sortedF[2][1]]
    return points
```
