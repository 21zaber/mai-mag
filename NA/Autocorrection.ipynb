# Автокорреляционный анализ

Автокорреляция - это мера тесноты и направления линейной стохастической зависимости между текущими значениями временного ряда и значениями временного ряда на $k$ моментов времени назад. Другими словами, автокорреляция - это «чистая корреляция» между $x_t$ и $x_{t-k}$.

Реализация:

```python
def autocorrelation(l):
    resp = []
    m = mean(l)
    n = len(l)
    
    for i in range(n):
        c = 0
        for j in range(n - i):
            c += (l[j] - m) * (l[i + j] - m)
        resp.append(c / (n - i))
        
    return resp
```

Несколько примеров использования 

```python
n = 1000
w = (2 * pi / n)

signal = array([sin(8 * w * t) + 2 * cos(3 * w * t) for t in range(n)])
plt.plot(arange(n), signal, 'r')

a = autocorrelation(signal)
plt.plot(arange(n), a, 'b')

plt.grid(True)
plt.show()
```

```python
n = 30
G = [random.gauss(0,1) for i in range(n)]
plt.plot(arange(n), G, 'ro')

a = autocorrelation(G)
plt.plot(arange(n), a, 'b')

plt.grid(True)
plt.show()
```
