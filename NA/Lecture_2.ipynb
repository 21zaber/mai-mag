# "хищник-жертва" и фазовый портрет

```python

import matplotlib.pyplot as plt
import numpy as np

def feval(funcName, *args):
    return eval(funcName)(*args)

def RK4(func, yinit, x_range, h):
    m = len(yinit)
    n = int((x_range[-1] - x_range[0])/h)

    x = x_range[0]
    y = yinit

    # Containers for solutions
    xsol = np.empty(0)
    xsol = np.append(xsol, x)

    ysol = np.empty(0)
    ysol = np.append(ysol, y)

    for i in range(n):
        k1 = feval(func, x, y)
        yp2 = y + k1*(h/2)
        k2 = feval(func, x+h/2, yp2)
        yp3 = y + k2*(h/2)
        k3 = feval(func, x+h/2, yp3)
        yp4 = y + k3*h
        k4 = feval(func, x+h, yp4)

        for j in range(m):
            y[j] = y[j] + (h/6)*(k1[j] + 2*k2[j] + 2*k3[j] + k4[j])

        x += h
        xsol = np.append(xsol, x)

        for r in range(len(y)):
            ysol = np.append(ysol, y[r])

    return [xsol, ysol]

def ODES(x, y):
    al1 = 1.5
    al2 = 1
    g1 = 0.8
    g2 = 0.4
    dy = np.zeros((len(y)))
    #dy[0] = -y[0] * al + g1 * y[0] * y[1];
    #dy[1] = -g1 * y[0] * y[1
    dy[0] = -y[0] * al1 + g1 * y[0] * y[1] +y[1]**2;
    dy[1] = y[1] * al2 - g2 * y[0] * y[1];
    return dy

h = 0.01
x = np.array([0.0, 30.0])
yinit = np.array([10.0, 2.0])

[ts, ys] = RK4('ODES', yinit, x, h)

node = len(yinit)
ys1 = ys[0::node]
ys2 = ys[1::node]
```

```python
def trapezoidal(f, a, b, n):
    h = (b - a)/n
    result = 0.5*(f[0] + f[n - 1])
    for i in range(1, n):
        result += f[i]
    result *= h
    return result
```
